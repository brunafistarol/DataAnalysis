```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

# Processamento de dados

```python
df_hotel = pd.read_csv('hotel_bookings.csv')
```

```python
df_hotel.info()
```

```python
df_hotel.describe()
```

Verificando os valores de cada coluna:

```python
for i in df_hotel.columns:
    print(i, df_hotel[i].unique())
```

O output anterior mostra que existem linhas com zero adultos, então provavelmente há linhas sem hóspedes. Há também linhas onde o número de crianças é igual a NaN, o que precisa ser substituído para poder operar com a soma das colunas de hóspedes

Substituindo valores NaN e removendo linhas sem hóspedes:

```python
df_hotel['children'] = df_hotel['children'].fillna(0)
df_hotel['total_guests'] = df_hotel['adults'] + df_hotel['children'] + df_hotel['babies']
df_hotel = df_hotel.drop(df_hotel[df_hotel['total_guests'] == 0].index)
```

Precisamos analisar separadamente as informações de cada hotel, uma vez que são diferentes:

```python
df_hotel_resort = df_hotel.loc[(df_hotel["hotel"] == "Resort Hotel")]
df_hotel_city = df_hotel.loc[(df_hotel["hotel"] == "City Hotel")]
```

# Características, preferências e contagem dos hóspedes

Com relação a características e preferências dos hóspedes, apêndice fornecido informa que há informação disponível a respeito de:
 - faixa de idade dos hóspedes (adulto, criança ou bebê)
 - o país de origem
 - por onde o hóspede fez a reserva
 - refeições incluídas na reserva
 - tipo de quarto reservado
 - tipo de consumidor

### Faixa de idade dos hóspedes

```python
guests_age_city = df_hotel_city[df_hotel_city['is_canceled'] == 0][['adults', 'children', 'babies']].sum()/sum(df_hotel_city[df_hotel_city['is_canceled'] == 0]['total_guests'])
guests_age_resort = df_hotel_resort[df_hotel_resort['is_canceled'] == 0][['adults', 'children', 'babies']].sum()/sum(df_hotel_resort[df_hotel_resort['is_canceled'] == 0]['total_guests'])
```

```python
print(guests_age_city)
print(guests_age_resort)
```

```python
plt.bar([1, 3, 5], guests_age_city, label = 'City')
plt.bar([2, 4, 6], guests_age_resort, label = 'Resort')
plt.xticks([1.5, 3.5, 5.5], ['Adultos', 'Crianças', 'Bebês'])
plt.ylabel('Percentual')
plt.title('Percentual de faixas etárias nos hotéis', fontweight ='bold', fontsize = 15)
plt.rcParams["figure.figsize"] = (10,6)
plt.legend()
plt.show()
```

```python
plt.bar([1, 2, 3, 4, 5], guests_country_city.head(5))
plt.xticks([1, 2, 3, 4, 5], list(zip(*guests_country_city.head(5).index))[0])
plt.ylabel('Percentual')
plt.title('Percentual de hóspedes por país de origem no hotel da cidade - top 5', fontweight ='bold', fontsize = 14)
plt.rcParams["figure.figsize"] = (10,8)
plt.show()
```

### Tamanho médio da estadia

```python
stay_city = df_hotel_city[(df_hotel_city['is_canceled'] == 0)][['length_stay', 'arrival_date_year', 'arrival_date_month']]
stay_resort = df_hotel_resort[(df_hotel_resort['is_canceled'] == 0)][['length_stay', 'arrival_date_year', 'arrival_date_month']]
```
